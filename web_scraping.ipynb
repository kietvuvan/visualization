```python
def scrap_web(str):
  url="https://www.advfn.com/nyse/newyorkstockexchange.asp?companies="+str
  page=re.get(url)
  bs=BeautifulSoup(page.text,'html.parser')
  row_ts0=bs.find_all('tr',attrs={'class':'ts0'})
  row_ts1=bs.find_all('tr',attrs={'class':'ts1'})

  for i in row_ts0:
    row=i.find_all('td')
    company_names.append(row[0].text.strip())
    company_ticket.append(row[1].text.strip())

  for j in row_ts1:
    row=j.find_all('td')
    company_names.append(row[0].text.strip())
    company_ticket.append(row[1].text.strip())

  return company_names,company_ticket
```

```python

lis=['A','B','C','D','E','F','G','H','I','J','K']
for str in lis:
  names,ticket=scrap_web(str)
```

```python
# create data frame
df=pd.DataFrame(columns=['company_name','company_ticket'])
df['company_name']=names
df['company_ticket']=ticket
```
